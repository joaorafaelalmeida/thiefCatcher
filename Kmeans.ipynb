```python
from sklearn.cluster import KMeans
from numpy import sqrt, random, array, argsort
from sklearn.preprocessing import scale
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
```

```python
random.seed(123)
def makeData(N):
	x = []
	for i in range(N):
		a = i/1000 + random.uniform(-3, 2)
		r = random.uniform(-5, 10)
		if(r >= 9.9):
			r = r + 10
		elif(r<(-4.8)):
			r = r +(- 10)			
		x.append([a + r])	
	return array(x)

x = makeData(500)
```

```python
x_ax = range(500)
plt.plot(x_ax, x)
plt.show()
```

```python
x = scale(x)
```

```python
kmeans = KMeans(n_clusters = 1).fit(x)
print(kmeans)
```

```python
center = kmeans.cluster_centers_
print(center)
```

```python
distance = sqrt((x - center)**2)
```

```python
order_index = argsort(distance, axis = 0)
indexes = order_index[-5:]
```

```python
values = x[indexes]
```

```python
plt.plot(x_ax, x)
plt.scatter(indexes, values, color='r')
plt.show()
```

```python
boston = load_boston()
y =  boston.target
y = y.reshape(y.shape[0], 1)
y = scale(y)
```

```python
kmeans = KMeans(n_clusters = 1).fit(y)
print(kmeans)

center = kmeans.cluster_centers_
print(center)

distance = sqrt((y - center)**2)
order_index = argsort(distance, axis = 0)
indexes = order_index[-10:]
values = y[indexes]
```

```python
x_ax = range(y.shape[0])
plt.plot(x_ax, y)
plt.scatter(indexes,values, color='r')
plt.show()
```
