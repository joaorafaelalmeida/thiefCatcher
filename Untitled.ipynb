```python
def extractStats(data):
    nSamp,nCols=data.shape

    M1=np.mean(data,axis=0) # por cada coluna = 4
    Md1=np.median(data,axis=0)
    Std1=np.std(data,axis=0)
    #S1=stats.skew(data)
    ##K1=stats.kurtosis(data)
    #p=[75,90,95,98]
    #Pr1=np.array(np.percentile(data,p,axis=0)).T.flatten()
        
    #features=np.hstack((M1,Md1,Std1,S1,Pr1))
    features=np.hstack((M1,Md1,Std1))
    return(features)

def extratctSilenceActivity(data,threshold=0):
    if(data[0]<=threshold):
        s=[1]
        a=[]
    else:
        s=[]
        a=[1]
    for i in range(1,len(data)):
        if(data[i-1]>threshold and data[i]<=threshold):
            s.append(1)
        elif(data[i-1]<=threshold and data[i]>threshold):
            a.append(1)
        elif (data[i-1]<=threshold and data[i]<=threshold):
            s[-1]+=1
        else:
            a[-1]+=1
    return(s,a)
    
def extractStatsSilenceActivity(data):
    #REMOVER BYTES
    features=[]
    nSamples,nMetrics=data.shape
    silence_features=np.array([])
    activity_features=np.array([])
    for c in range(nMetrics):
        silence,activity=extratctSilenceActivity(data[:,c],threshold=0)
        
        if len(silence)>0:
            silence_faux=np.array([len(silence),np.mean(silence),np.std(silence)])
        else:
            silence_faux=np.zeros(3)
        silence_features=np.hstack((silence_features,silence_faux))
        
        if len(activity)>0:
            activity_faux=np.array([len(activity),np.mean(activity),np.std(activity)])
        else:
            activity_faux=np.zeros(3)
        activity_features=np.hstack((activity_features,activity_faux))	   
    features=np.hstack((silence_features,activity_features))
    return(features)
```

### slidingMultObsWindow
This function returns all the extracted features. The return is a dict containing the windows size (as key) and the list of features extracted from that windows (as value).
Index of the features in the array:
1. Mean (4 columns: Up Size, Up Count, Down Size, Down Count) [0, 1, 2, 3]
2. Median (4 columns: Up Size, Up Count, Down Size, Down Count) [4, 5, 6, 7]
3. Standard deviation (4 columns: Up Size, Up Count, Down Size, Down Count) [8, 9, 10, 11]
4. Silence (12 columns)
5. Activity (12 columns)

```python
def slidingMultObsWindow(data,lengthObsWindow,slidingValue, filterIndex=None):
    nSamples,nMetrics=data.shape
    #Fix windows
    results = {}
    for s in np.arange(max(lengthObsWindow),nSamples,slidingValue):
        features=np.array([])
        for oW in lengthObsWindow:
            if oW not in results:
                results[oW] = []
            subdata=data[s-oW:s,1:]
            faux=extractStats(subdata)
            faux2=extractStatsSilenceActivity(subdata)
            features=np.hstack((faux,faux2))#np.hstack((features,faux,faux2))
            #print(('{} '*len(features)).format(*features))
            if filterIndex:
                features = features[filterIndex]
            results[oW].append(features)
    
    for oW in lengthObsWindow:
        if oW in results:
            results[oW] = np.array(results[oW])
    return results
```

```python

fileInput="results/normal/tcpWindows17_12_2020_19_59_38.txt"
bruteInput="results/brute/tcpWindows17_12_2020_21_44_24.txt"
smoothInput="results/smooth/tcpWindows17_12_2020_21_59_40.txt"
intelligentInput="results/intelligent/tcpWindows17_12_2020_23_43_16.txt"

window=8
obsWindows=[8]#,32,64]#,10]
slidingValue=4
featuresIndex=[2, 10]

data=np.loadtxt(fileInput,dtype=int)
normalTraffic = slidingMultObsWindow(data,obsWindows,slidingValue, featuresIndex)

dataBrute=np.loadtxt(bruteInput,dtype=int)
bruteTraffic = slidingMultObsWindow(dataBrute,obsWindows,slidingValue, featuresIndex)

dataSmooth=np.loadtxt(smoothInput,dtype=int)
smoothTraffic = slidingMultObsWindow(dataSmooth,obsWindows,slidingValue, featuresIndex)

dataIntelligent=np.loadtxt(intelligentInput,dtype=int)
intelligentTraffic = slidingMultObsWindow(dataIntelligent,obsWindows,slidingValue, featuresIndex)

```
